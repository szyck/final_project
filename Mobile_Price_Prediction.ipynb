```python
#Set inputs for training model
X = df_train.drop('price_range',axis=1)
```

```python
X.head()
```

```python
#Set output for training model
Y=df_train['price_range']
```

```python
Y
```

```python
#Reshape the testing model
df_test=df_test.drop('id',axis=1)
```

```python
df_test.shape
```

```python
from sklearn.preprocessing import StandardScaler
std=StandardScaler()
```

```python
#Scale the training model
X_std=std.fit_transform(X)
```

```python
X_std
```

```python
#Scale the testing model
df_test_std=std.transform(df_test)
```

```python
df_test_std
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=1)
X_train
```

```python
Y_train
```

```python
X_test
```

```python
# SET UP DECISION TREE:
from sklearn.tree import DecisionTreeClassifier
dt=DecisionTreeClassifier()
```

```python
#Fit the training model
dt.fit(X_train,Y_train)
```

```python
#Apply triaing model to test data
Y_pred=dt.predict(X_test)
```

```python
#Predicted outcomes per the model
Y_pred
```

```python
#Test accuracy of model
from sklearn.metrics import accuracy_score
dec_tree_accuracy=accuracy_score(Y_test,Y_pred)
dec_tree_accuracy
```
